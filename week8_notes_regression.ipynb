# Regression, general

Regression analysis focuses on the relationship between variables. 
For example, does the price of a house depend on the number of bedrooms or the square feet area.

The dependent feature (house price) is called the dependent variable or response variable, usually written as y.

The independent feature (number of bedroom) is called the independent variable or predictor variable, usually written as x.

Regression analysis can be applied to both continuous abnd categorical data.

Regression can be used to determine how variables are related or to forcast a response.

Linear regression is a widely used regression technique. It assumes a linear relationship between the variables such that 

𝑦 = 𝛽₀ + 𝛽₁𝑥₁ + ⋯ + 𝛽ᵣ𝑥ᵣ + 𝜀.

where  𝛽₀, 𝛽₁, …, 𝛽ᵣ ... are the regression coefficients and 𝜀 is the random error. 

Linear regression calculates the estimators of the regression coefficients or the predicted weights, denoted with 𝑏₀, 𝑏₁, …, 𝑏ᵣ. The estimated regression function 𝑓(𝐱) = 𝑏₀ + 𝑏₁𝑥₁ + ⋯ + 𝑏ᵣ𝑥ᵣ describes the dependencies between the inputs and output.

The estimated or predicted response, 𝑓(𝐱ᵢ), for each observation 𝑖 = 1, …, 𝑛, should be as close as possible to the corresponding actual response 𝑦ᵢ. The differences 𝑦ᵢ - 𝑓(𝐱ᵢ) for all observations 𝑖 = 1, …, 𝑛, are called the residuals. Regression is applied to capture the weights corresponding to the smallest residuals.

One way to obtain the best weights is to minimize the sum of squared residuals (SSR) for all observations 𝑖 = 1, …, 𝑛: SSR = Σᵢ(𝑦ᵢ - 𝑓(𝐱ᵢ))². This approach is called the method of ordinary least squares.

# Simple linear regression

Simple or single-variate linear regression is linear regression with a single independent variable, 𝐱 = 𝑥. Multiple regression will be addressed elsewhere.


When working with simple linear regression, a given set of input-output (𝑥-𝑦) pairs  represents the  observations. 

For example, above, Measure B, or x = 10 and Measure A, or y = 9.

The estimated regression function is:
 𝑓(𝑥) = 𝑏₀ + 𝑏₁𝑥.

The goal is to calculate the optimal values of the predicted weights 𝑏₀ and 𝑏₁ that minimize SSR (sum of squared residuals) and determine the estimated regression function.

*The value of 𝑏₀, also called the intercept, shows the point where the estimated regression line crosses the 𝑦 axis.*
 It is the value of the estimated response 𝑓(𝑥) for 𝑥 = 0.* The value of 𝑏₁ determines the slope of the estimated regression line.*

The predicted responses are the points on the regression line that correspond to the input values. For example, for the input 𝑥 = 5, the predicted response is 𝑓(10) = 9.
One way to visualize the data is to use a scatter plot with a trend line,which you can do with the seaborn regplot() function. 
The residuals, represented by vertical  lines can be calculated as 𝑦ᵢ - 𝑓(𝐱ᵢ) = 𝑦ᵢ - 𝑏₀ - 𝑏₁𝑥ᵢ for 𝑖 = 1, …, 𝑛. They represent the distances between the actual response and the predicted response. The goal is to minimize these distances.

```python
# example
#linear regression least square implementation

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#customize figsize
plt.rcParams['figure.figsize'] = (10.0, 5.0)

# read data
data = pd.read_csv('https://raw.githubusercontent.com/ava11235/it125/main/datasets/headbrain.csv')

print(data.shape)
print(data.head())

# collect X and Y values
X = data['Head Size(cm^3)'].values
Y = data['Brain Weight(grams)'].values

# calculate means Y
mean_x = np.mean(X)
mean_y = np.mean(Y)

print("Head size mean:", mean_x)
print("Brain weight mean:", mean_y)

# number of values
n = len(X)

# calculate m and c
numer = 0
denom = 0
for i in range(n):
    numer += (X[i] - mean_x) * (Y[i] - mean_y)
    denom += (X[i] - mean_x) ** 2
    m = numer / denom
    c = mean_y - (m * mean_x)
 
# print coefficients
print("coefficients:",  m, c)

# plot values and regression line
max_x = np.max(X) + 100
min_x = np.min(X) - 100

# calculate line values for x and y
x = np.linspace(min_x, max_x, 1000)
y = c + m * x 
 
# plot Line
plt.plot(x, y, color='red', label='Regression Line')

# plot scatter points
plt.scatter(X, Y, c='purple', label='Scatter Plot')
 
plt.xlabel('Head Size in cm3')
plt.ylabel('Brain Weight in grams')
plt.legend()
plt.show()

#plt.savefig("regline.png")
```
