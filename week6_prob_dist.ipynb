**Probability Distributions with Python**

A random variable is a variable whose possible outcomes are random. There are discrete and continous random variables.

A discrete random variable is a variable whose values can only be distinct, countable values, such as 1,2,3,4,5.

The probability distribution of a discrete random variable, also called probability mass function, is a list of probabilities associated with each of its possible values.

Examples of discrete probability distributions include the Bernoulli and the Poisson distributions.

A continuous random variable is one which takes an infinite number of possible values. For example, a random variable defined to be the height of students in a class, is a continous random variable. A continuous random variable is represented by the area under a curve, since it is defined over an interval of values.

The probability distribution of a continuous random variable is known as a probability distribution function where the function can take on continuous values. Since the number of values this type of random variable can take is infinite, the probability of observing any single value is equal to 0. For example, a random variable may be assigned all values over the interval of real numbers.

An example of a continuous probability distribution is the Normal distribution.

The **scipy.stats module** contains functions to work with probability distributions and statistical functions.

Use **Seaborn’s displot** function to plot the histogram of distributions. Seaborn’s displot has multiple arguments to customize the plot such as the number of bins, the color of the histogram  or the kde option. Use the axis .set to set the  x and y axis labels.

# Binomial distribution
A Binomial distribution is a distribution with only two possible outcomes, such as heads/tails, or success/failure. The probability of both outcomes is the same for all trials. Note that the outcomes need not be equally likely. Each trial is independent. The parameters of a binomial distribution are n and p where n is the total number of trials, and p is the probability of success in each trial.The Bernoulli distribution is a special case of the binomial distribution where a single trial is conducted.

You can generate a binomial distributed discrete random with the scipy.stats module's binom.rvs() method which takes n, number of trials and p, probability of success as arguments. The size argument determines the number of times the trails should be repeated. 

*In this and other examples, if you want to maintain reproducibility, include a random_state argument assigned to a number.



```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from scipy.stats import binom
data_binom = binom.rvs(n=100,p=0.7,size=10000, random_state=5)
ax = sns.displot(data_binom, color="salmon")
ax.set(xlabel='Binomial Distribution', ylabel='Frequency')
```

# Poisson Distribution

A Poisson random variable can model the number of times an event happened during  a time interval. For example, the count of a website visitors during a sale event (time interval) can be expressed as a Poisson process. Poisson distribution is described by the rate (μ) at which the events happen. An event can occur 0, 1, 2, etc times during an interval. The average number of events in an interval is designated as λ (lambda). This is also known as the event rate, or rate parameter. 

Use the scipy.stats module's poisson.rvs() method which takes μ as a shape parameter and the  λ in the equation. The size decides the number of random variates in the distribution. Visualize the distribution with the use of seaborn's dislot function.


```python
from scipy.stats import poisson
data_poisson = poisson.rvs(mu=5, size=10000)
ax = sns.displot(data_poisson, bins=30, color="aquamarine" )
ax.set(xlabel='Poisson Distribution', ylabel='Frequency')
```

# Normal distribution
A normal distribution, also known as Gaussian distribution, is commonly encountered in Data Science and Analysis.

A normal distribution has a bell-shaped density curve described by its mean μ and standard deviation σ. The density curve is symmetrical and centered around the mean. The spread of this distribution is determined by its standard deviation showing that data near the mean occur more frequently.

With the Normal Distribution Function almost 68% of its data falls within one standard deviation distance from the mean and on either side of it. 95% of its data fall within two standard deviations from the mean. 

To generate a normally distributed random variable use the scipy.stats module's norm.rvs() method. The loc argument is the mean of the distribution. scale is the standard deviation, and size to the number of random variates. To visualize the normal distribution, use seaborn's displot function.



```python
from scipy.stats import norm
# generate random numbers from N(0,1)
data_normal = norm.rvs(size=10000,scale=1)
ax = sns.displot(data_normal, bins=100, kde=True, color='r')
ax.set(xlabel='Normal Distribution', ylabel='Frequency')
oc=0
```
